```python
def fun(x):
    return (x+3)**2
```

```python
def der(x):
    return 2*(x+3)
```

```python
def gred(st,alpha,iter):
    xl=list()
    x=st
    xl.append(x)
    for i in range(iter):
        x=x-alpha*der(x)
        xl.append(x)
    return xl
```
